## 2. A review on Gaussian likelihood
Let's start from the basics. Remember what a likelihood is.
The **likelihood** measures the goodness of fit of a statistical model to samples of data for given values of
the unknown model parameters.
It's computed from the joint probability distribution, but viewed and used as **function**
of the parameters only, thus treating the random variables as fixed at the observed values.

A Gaussian likelihood is defined as

\begin{equation}
p(\mathbf{y}|\mathbf{w}, \mathbf{X}, \sigma_\mathrm{n}) = \prod_{i=1}^N p(y_i|\mathbf{w}, {\mathbf{x}}_i, \sigma_\mathrm{n}) = \prod_{i=1}^N \mathcal{N}(y_i|\tilde y_i, \sigma_\mathrm{n})
\end{equation}

where, for linear regression, $\tilde y_i = \mathbf{w}^\top {\mathbf{x}}_i$.
For numerical stability, instead of using the vanilla likelihood, we will use the **log-likelihood**.

\begin{equation}
\log p(\mathbf{y}|\mathbf{w}, \mathbf{X}, \sigma_\mathrm{n}) =  \sum_{i=1}^N  \log\mathcal{N}(y_i|\tilde y_i, \sigma_\mathrm{n})
\end{equation}

```python
import numpy as np
import matplotlib.pyplot as plt

def lognormal(x, mu, var):
    """Compute log-density of normal distribution"""
    return -0.5 * np.log(2 * np.pi * var) - 0.5 * ((x - mu)**2 / var)

def gaussian_loglik(y, y_tilde, sn2):
    """Compute log-likelihood for a single sample"""
    return lognormal(y, y_tilde, sn2)

def plot_gaussian(mu, var, plot_log=False, **kwargs):
    """Plot a gaussian pdf"""
    x = np.linspace(mu - 5 * np.sqrt(var), mu + 5 * np.sqrt(var), 200)
    y = lognormal(x, mu, var) if plot_log else np.exp(lognormal(x, mu, var))
    ax = kwargs.pop("ax", plt.gca())
    ax.plot(x, y, **kwargs)
    return ax

y_obs = 0.4
y_tilde = 0.3
sn2 = 1.0

sample_ll = gaussian_loglik(y_obs, y_tilde, sn2)

fig, ax = plt.subplots(figsize=[6, 4])

plot_gaussian(y_tilde, sn2, ax=ax, linewidth=2,
             label=f'Gaussian distribution\n centered at prediction $\widetilde{{y}} = {y_tilde}$')

ax.vlines(y_obs, 0, np.exp(sample_ll), color='red', linewidth=2,
         label=f'Observed value $y = {y_obs}$')

ax.vlines(y_tilde, 0, np.exp(gaussian_loglik(y_tilde, y_tilde, sn2)),
         ls='--', color='blue', alpha=0.5, linewidth=1.5,
         label=f'Predicted value $\widetilde{{y}} = {y_tilde}$ (peak)')

ax.plot(y_obs, np.exp(sample_ll), 'ro', markersize=8,
       label=f'Likelihood = {np.exp(sample_ll):.3f}')

ax.text(y_tilde + 0.1, 0.05, f'Predicted value\n$\widetilde{{y}} = {y_tilde}$',
        fontsize=9, bbox=dict(facecolor='white', alpha=0.8))
ax.text(y_obs - 0.4, 0.12, f'Observed value\n$y = {y_obs}$',
        fontsize=9, bbox=dict(facecolor='white', alpha=0.8))

ax.set_xlabel('$y$ (observation)', fontsize=11)
ax.set_ylabel('Probability Density', fontsize=11)
ax.set_title('Gaussian Likelihood: How likely is the observed value\n given our prediction?',
             fontsize=12, fontweight='bold')

ax.grid(True, alpha=0.3)

ax.set_ylim(bottom=0)

ax.legend(loc='upper left', fontsize=9)

ax.annotate('The likelihood is the height of the\nGaussian curve at the observed value',
           xy=(y_obs, np.exp(sample_ll)), xytext=(1.5, 0.25),
           arrowprops=dict(arrowstyle='->', color='gray'),
           bbox=dict(boxstyle='round,pad=0.3', facecolor='yellow', alpha=0.7),
           fontsize=9)

plt.tight_layout()
plt.show()

print("\n" + "="*60)
print("SIMPLE EXPLANATION:")
print("="*60)
print(f"• Our model PREDICTS the value should be: {y_tilde}")
print(f"• But we OBSERVED the value: {y_obs}")
print(f"• We're not certain - there's noise with variance: {sn2}")
print(f"• The Gaussian curve shows all possible values we might observe")
print(f"• The LIKELIHOOD asks: 'Given our prediction, how probable is it")
print(f"  to observe exactly {y_obs}?'")
print(f"• Answer: The likelihood is {np.exp(sample_ll):.3f}")
print(f"• This is the HEIGHT of the curve at y = {y_obs}")
print("="*60)
```

## 3. Bayesian Linear regression

In this section, you'll start to implement the Bayesian linear regression model.
Let's start by creating the **design matrix** $\mathbf{X}$.

$$
\mathbf{X} = \left[ {\begin{array}{ccccc}
   1 & x_1^1 & \dots & x_1^K\\
   1 & x_2^1 & \dots & x_2^K\\
   \vdots &    \vdots & &   \vdots \\
   1 & x_N^1 & \dots & x_N^K\\
  \end{array} } \right]
$$

In any case, inspect $\mathbf{X}$ to make sure it looks OK (show the first entries). To fit higher order polynomials, we need to add extra columns to $\mathbf{X}$, therefore build it with $K$ as a parameter.

```python
import numpy as np
import matplotlib.pyplot as plt

# First, let's create some sample data to work with (using the dataset from earlier)
np.random.seed(1)
true_function = lambda x: 0.5 * ((x - 1) ** 2) - 3

def make_regression(n, sn2=0.1):
    x = np.random.uniform(-3, 3, n)
    y = true_function(x) + np.sqrt(sn2) * np.random.randn(*x.shape)
    return x, y

# Generate the dataset
sn2 = 1.5
x, y = make_regression(30, sn2=sn2)

def build_features(x, K):
    """
    Build the design matrix X for polynomial regression of degree K.

    Parameters:
    x: input data (1D array)
    K: polynomial degree (number of features = K+1 including bias term)

    Returns:
    X: design matrix of shape (N, K+1)
    """
    # Method 1: Using numpy column_stack (recommended)
    X = np.column_stack([x**i for i in range(K+1)])

    # Alternative Method 2: Using list comprehension
    # X = np.array([[xi**i for i in range(K+1)] for xi in x])

    return X

# Test the function with K=2 (quadratic)
X = build_features(x, 2)
print("Design matrix X (first 5 rows):")
print(X[:5])
print(f"\nShape of X: {X.shape}")
print(f"This means: {X.shape[0]} samples, {X.shape[1]} features (bias + x + x²)")
```

```python
def compute_posterior(X, y, sigma_n, sigma_w):
    """
    Compute the posterior distribution of weights w given data.

    Parameters:
    X: design matrix (N x M)
    y: target values (N,)
    sigma_n: noise standard deviation
    sigma_w: prior standard deviation for weights

    Returns:
    mu_w: posterior mean of weights
    Sigma_w: posterior covariance matrix of weights
    """
    N, M = X.shape

    # Prior precision matrix (inverse of covariance)
    S_inv = (1/sigma_w**2) * np.eye(M)

    # Noise precision
    noise_precision = 1/sigma_n**2

    # Compute posterior covariance: Σ^{-1} = (1/σ_n²) X^T X + S^{-1}
    Sigma_w_inv = noise_precision * X.T @ X + S_inv

    # Solve for Sigma_w (instead of computing inverse directly)
    # This is more numerically stable
    Sigma_w = np.linalg.inv(Sigma_w_inv)  # For small problems, inverse is ok

    # Compute posterior mean: μ_w = (1/σ_n²) Σ X^T y
    mu_w = noise_precision * Sigma_w @ X.T @ y

    return mu_w, Sigma_w

# Set parameters
sigma_n = np.sqrt(sn2)  # noise standard deviation
sigma_w = 1.0           # prior standard deviation for weights

# Compute posterior
mu_w, Sigma_w = compute_posterior(X, y, sigma_n, sigma_w)

print("Posterior mean of weights (w):")
print(mu_w)
print(f"\nPosterior covariance matrix shape: {Sigma_w.shape}")
print(f"This is a {Sigma_w.shape[0]}x{Sigma_w.shape[1]} matrix")
```

From the lecture notes, let's define the prior on the parameters $\mathbf{w}$ as

\begin{equation}
\mathbf{w} \sim \mathcal{N}(\mathbf{0}, \mathbf{S})
\end{equation}

For sake of simplicity, assume the covariance matrix $\mathbf{S}$ to be diagonal $\mathbf{S} = \sigma_\mathrm{w}^2\mathbf{I}$.
Remember that the likelihood is defined as $p(\mathbf{y}|\mathbf{w}, \mathbf{X}, \sigma_\mathrm{n}) = \mathcal{N}(\mathbf{y}|\mathbf{X}\mathbf{w}, \sigma_\mathrm{n}^2\mathbf{I})$.
In this case, the posterior is analitic and follows this form:

\begin{equation}
p(\mathbf{w}|\mathbf{X}, \mathbf{y}, \sigma_\mathrm{n}) = \mathcal{N}\left(\frac{1}{\sigma^2_\mathrm{n}}\boldsymbol{\Sigma}\mathbf{X}^\top\mathbf{y}, \boldsymbol{\Sigma} \right)
\end{equation}
where $\boldsymbol{\Sigma}^{-1} = \left(\frac{1}{\sigma^2_\mathrm{n}}\mathbf{X}^\top\mathbf{X} + \mathbf{S}^{-1}  \right)$.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(1)
true_function = lambda x: 0.5 * ((x - 1) ** 2) - 3

def make_regression(n, sn2=0.1):
    x = np.random.uniform(-3, 3, n)
    y = true_function(x) + np.sqrt(sn2) * np.random.randn(*x.shape)
    return x, y

sn2 = 1.5
x, y = make_regression(30, sn2=sn2)

def build_features(x, K):
    X = np.column_stack([x**i for i in range(K+1)])
    return X

K = 2
X = build_features(x, K)

def compute_posterior(X, y, sigma_n, sigma_w):
    N, M = X.shape
    S_inv = (1/sigma_w**2) * np.eye(M)
    beta = 1/sigma_n**2
    Sigma_w_inv = beta * X.T @ X + S_inv
    Sigma_w = np.linalg.inv(Sigma_w_inv)
    mu_w = beta * Sigma_w @ X.T @ y
    return mu_w, Sigma_w

sigma_n = np.sqrt(sn2)
sigma_w = 1.0

mu_w, Sigma_w = compute_posterior(X, y, sigma_n, sigma_w)

print("Posterior mean of weights (w):")
print(mu_w)
print(f"\nPosterior covariance matrix shape: {Sigma_w.shape}")

# ============================================
# ANSWERS TO QUESTIONS
# ============================================
print("\n" + "="*60)
print("ANSWERS TO QUESTIONS")
print("="*60)

M = X.shape[1]
print(f"1. Dimensionality of Σ: {M} x {M}")
print(f"   (Σ is a {M}x{M} matrix, where M = {M} is the number of features)")

print(f"\n2. Computational cost to compute Σ⁻¹:")
print(f"   - Computing XᵀX: O(N × M²) operations")
print(f"   - Inverting the {M}x{M} matrix: O(M³) operations")
print(f"   - Total cost: O(N × M² + M³)")
print(f"   For our case: N={len(x)}, M={M} → approximately {len(x)*M**2 + M**3} operations")

print(f"\n3. For numerical stability, instead of computing inverse directly:")
print(f"   - Use Cholesky decomposition (for symmetric positive definite matrices)")
print(f"   - Or solve linear systems directly using np.linalg.solve()")
print(f"   - Never compute A⁻¹ explicitly when you only need A⁻¹b")

# ============================================
# NUMERICALLY STABLE ALTERNATIVE (FIXED)
# ============================================
print("\n" + "="*60)
print("NUMERICALLY STABLE ALTERNATIVE")
print("="*60)

def compute_posterior_stable(X, y, sigma_n, sigma_w):
    """More numerically stable version using Cholesky decomposition"""
    N, M = X.shape


    S_inv = (1/sigma_w**2) * np.eye(M)
    beta = 1/sigma_n**2

    A = beta * X.T @ X + S_inv
    L = np.linalg.cholesky(A)


    b = beta * X.T @ y
    z = np.linalg.solve(L, b)
    mu_w = np.linalg.solve(L.T, z)

    I = np.eye(M)
    Y = np.linalg.solve(L, I)
    Sigma_w = np.linalg.solve(L.T, Y)

    return mu_w, Sigma_w

mu_w_stable, Sigma_w_stable = compute_posterior_stable(X, y, sigma_n, sigma_w)

print(f"\nPosterior mean (stable method): {mu_w_stable}")
print(f"Posterior mean (original method): {mu_w}")
print(f"\nPosterior covariance (stable method):\n{Sigma_w_stable}")
print(f"\nPosterior covariance (original method):\n{Sigma_w}")

print(f"\nMeans are equal: {np.allclose(mu_w, mu_w_stable)}")
print(f"Covariances are equal: {np.allclose(Sigma_w, Sigma_w_stable)}")
```

## 7. A more complex model: being Bayesian on the noise

Now we are ready spice things up! Everything we did, we did it assuming that the noise variance $\sigma_\mathrm{n}^2$ was known. This is a far too restrictive assumption in practice.
There are several ways to choose $\sigma_\mathrm{n}^2$ (cross-validation, maximization of the marginal likelihood w.r.t. $\sigma_\mathrm{n}^2$, ...).
But this -- being a course on Bayesian inference -- requires to develop a solution in a "Bayesian" way: place a prior on $\sigma_\mathrm{n}^2$ and infer a posterior given some data.

As usual, the likelihood has the form $p(\mathbf{y}|\mathbf{w}, \mathbf{X}, \sigma_\mathrm{n}) = \mathcal{N}(\mathbf{X}\mathbf{w}, \sigma^2_\mathrm{n} \mathbf{I})$.

One can show that the natural conjugate prior is a normal-inverse Gaussian distribution with the following form:

$$
p(\mathbf{w}, \sigma^2_\mathrm{n})  = \mathrm{NIG}(\mathbf{w}, \sigma^2_\mathrm{n}|\mu_0, \mathbf{\Sigma}_0, a_0, b_0) =
$$
$$
                                    = \mathcal{N}(\mathbf{w}|\mu_0,\sigma^2_\mathrm{n}\mathbf{\Sigma}_0)\mathrm{IG}(\sigma^2_\mathrm{n}|a_0, b_0)
                                    $$


where IG is the inverse Gamma distribution.

```python
a0, b0 = 1, 3
rv = scipy.stats.invgamma(a=a0, scale=b0)

fig, ax = plt.subplots(figsize=[4, 2.5])
xplot = np.linspace(0, 10, 100)
ax.plot(xplot, rv.pdf(xplot))
ax.set_title("IG(%.1f,%.1f)" % (a0, b0))
plt.show()
```

One can derive the posterior $p(\mathbf{w},\sigma_\mathrm{n}^2 | \mathbf{X}, \mathbf{y})$ as follows

\begin{align}
p(\mathbf{w},\sigma_\mathrm{n}^2 | \mathbf{X}, \mathbf{y}) &= \mathrm{NIG}(\mathbf{w},\sigma_\mathrm{n}^2|\mu, \mathbf{\Sigma}, a, b)\\
\mathbf{\Sigma} &= (\mathbf{\Sigma}_0^{-1} + \mathbf{X}^\top\mathbf{X})^{-1}\\
\mu &= \mathbf{\Sigma}(\mathbf{\Sigma}_0^{-1}\mu_0 + \mathbf{X}^\top\mathbf{y}) \\
a &= a_0 + n/2 \\
b &= b_0 + \frac{1}{2}(\mu_0^\top\mathbf{\Sigma}_0\mu_0 + \mathbf{y}^\top\mathbf{y} - \mu^\top\mathbf{\Sigma}\mu )
\end{align}

This is left as an exercise to the reader (joking aside if you want to know more check Sec. 7.6.3.1 of "Machine Learning: A probabilistic perspective" by K.P. Murphy).
From this formulation, you can derive the two marginals on $\mathbf{w}$ and ${\sigma_\mathrm{n}^2}$, which are easier to understand.

\begin{align}
p(\sigma_\mathrm{n}^2|\mathrm{X}, \mathrm{y}) &= \mathrm{IG}(a,b)\\
p(\mathbf{w}|\mathrm{X}, \mathrm{y}) &= \mathcal{T}\left(\mu, \frac{b}{a}\Sigma, 2a\right)
\end{align}

where $\mathcal{T}$ is a student-T distribution.

Finally, the predictive distribution a new test points $\mathbf{X}_{\mathrm{new}}$ is again a student-T distribution with the following form,

\begin{align}
p(\mathbf{y}_\mathrm{new}| \mathbf{X}_\mathrm{new}, \mathbf{X}, \mathbf{y}) = \mathcal{T}\left(\mathbf{X}_\mathrm{new}\mu, \frac{b}{a}(I + \mathbf{X}_\mathrm{new}\mathbf{\Sigma}\mathbf{X}_\mathrm{new}^\top), 2a \right)
\end{align}
